## Review Frames in Video

```python
seg = cv2.VideoCapture(str(segment_paths.video_paths[0]))

# Get the number of frames
frame_count = seg.get(cv2.CAP_PROP_FRAME_COUNT)

# Get the frames per second
fps = seg.get(cv2.CAP_PROP_FPS)

# Get the duration of the video
duration = frame_count / fps

print(f"Frame Count: {frame_count}")
print(f"FPS: {fps}")
print(f"Duration: {duration}")

# get the video width and height
width = int(seg.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(seg.get(cv2.CAP_PROP_FRAME_HEIGHT))

print(f"Width: {width}")
print(f"Height: {height}")
```

```python
def keyframe_detection_pixel_difference(video_path, threshold=0.2):
    """Identifies keyframes using pixel differences between frames.

    Args:
        video_path: Path to the video file.
        threshold: The threshold for determining a change (higher means fewer keyframes).

    Returns:
        list: A list of tuples, where each tuple is (frame, timestamp).
    """

    cap = cv2.VideoCapture(str(video_path))
    keyframes = []
    prev_frame = None
    frame_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if prev_frame is not None:
            diff = cv2.absdiff(frame, prev_frame)
            diff_sum = np.sum(diff)
            diff_norm = diff_sum / np.sqrt(np.sum(frame**2) * np.sum(prev_frame**2))

            if diff_norm > threshold:
                timestamp = cap.get(cv2.CAP_PROP_POS_MSEC) / 1000  # Timestamp in seconds
                keyframes.append((frame, timestamp))

        prev_frame = frame
        frame_count += 1

    cap.release()
    return keyframes

# def keyframe_detection_optical_flow(video_path, min_flow_change=5):
#     """Identifies keyframes using changes in optical flow.

#     Args:
#         video_path: Path to the video file.
#         min_flow_change: Minimum change in flow magnitude to consider a keyframe.

#     Returns:
#         list: A list of tuples, where each tuple is (frame, timestamp).
#     """

#     cap = cv2.VideoCapture(str(video_path))
#     keyframes = []
#     prev_frame = None
#     frame_count = 0

#     # Create an object for optical flow calculation (modify parameters as needed)
#     prev_gray = None
#     pyr_scale = 0.5
#     levels = 5 # 3
#     flow_params = dict(
#         pyr_scale=pyr_scale,
#         levels=levels,
#         winsize=(15, 15), 
#         iterations=3,
#         maxLevel=2, 
#         poly_n=5,
#         poly_sigma=1.2,
#         criteria=(cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03),
#     )

#     while True:
#         ret, frame = cap.read()
#         if not ret:
#             break

#         gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

#         if prev_gray is not None:
#             flow = cv2.calcOpticalFlowFarneback(prev_gray, gray, None, **flow_params)
#             mag, _ = cv2.cartToPolar(flow[..., 0], flow[..., 1])  # Calculate flow magnitude
#             avg_flow_change = np.mean(np.abs(mag))

#             if avg_flow_change > min_flow_change:
#                 timestamp = cap.get(cv2.CAP_PROP_POS_MSEC) / 1000  # Timestamp in seconds
#                 keyframes.append((frame.copy(), timestamp))  # Store a copy of the frame

#         prev_gray = gray
#         frame_count += 1

#     cap.release()
#     return keyframes
```
